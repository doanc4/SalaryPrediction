```python
df_text = df[['Title', 'ContentDescription']]
df_text['salary'] = salary
df_text = df_text.dropna()
df_text['ContentDescription'][15]
```

```python
rf.feature_importances_[123]
```

```python
stop_words = set(stopwords.words('english'))
def clean_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', ' ', string.punctuation))
    text = text.translate(str.maketrans('', '', string.digits))
    text_tokens = word_tokenize(text)
    text_tokens = [tok for tok in text_tokens if tok not in stop_words]
    text = ' '.join(text_tokens)
    text = re.sub('[^a-zA-Z0-9_\s]', '', text)
    return text
```

```python
df_text['ContentDescription'] = df_text['ContentDescription'].apply(clean_text)
```

```python
tfidf = TfidfVectorizer(min_df=2)
tfidf_vectors = tfidf.fit_transform(df_text['ContentDescription'])
```

```python
terms = tfidf.get_feature_names_out()
X = tfidf_vectors.todense()
tfidf_df = pd.DataFrame(X, columns=terms)
tfidf_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(tfidf_df, df_text['salary'], test_size=.3)

rf = RandomForestRegressor()
rf.fit(X_train, y_train)
preds = rf.predict(X_test)

r2_score(y_test, preds)
```

## bert document embeddings as features 

```python
df = pd.read_csv('DC_biotech_postings_May2020_to_July2021_subset.csv')
df = df.drop(['ID', 'PositionID', 'DateLastSeen', 'Date', 'Occupation2Code', 'Occupation2Name', 'Occupation6Code', 'Occupation6Name', 'Certifications'], axis=1)
df = df.drop(['LocationCity', 'LocationMSA', 'LocationCounty', 'OccupationFunction', 'Occupation8Code'], axis=1)
df = df.dropna()
df
```

```python
title = df['Title']
content = df['ContentDescription']
salary = df['SalaryAnnualEst']
df = df.drop(['Title', 'ContentDescription', 'SalaryAnnualEst'], axis=1)
df = df.reset_index(drop=True)
```

```python
df = pd.get_dummies(df)
df
```

```python
content = [clean_text(text) for text in content]
```

```python
tfidf = TfidfVectorizer(min_df=2)
X = tfidf.fit_transform(content)
features = tfidf.get_feature_names_out()
X = X.todense()
X = pd.DataFrame(X, columns=features)
X
```

```python
df = pd.concat([df, X], axis=1)
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, salary, test_size=.3, random_state=10)
```

```python
rf = RandomForestRegressor()

rf.fit(X_train, y_train)
```

```python
preds = rf.predict(X_test)
r2_score(y_test, preds)
```
